```python
import json
import numpy as np
from urllib.request import urlopen, Request

import matplotlib.pyplot as plt
from PIL import Image
import base64
import io
```

https://app.runwayml.com/models/bennnnnn6/StyleGANTwo_Nebular

```python
inputs = {
  "z": list(np.random.uniform(0, 10, 512)),
  "truncation": 0.8
}

req = Request(
  "https://stylegantwo-nebular-12766310.hosted-models.runwayml.cloud/v1/query",
  method="POST",
  headers={
    "Accept": "application/json",
    "Content-Type": "application/json",
  },
  data=json.dumps(inputs).encode("utf8")
)

with urlopen(req) as url:
  output = json.loads(url.read().decode("utf8"))

image = output["image"]
# use the outputs in your project
```

```python

img = Image.open(io.BytesIO(base64.b64decode(image[image.find('/9'):])))

plt.figure(figsize=(20,10))
plt.imshow(img)
plt.show()
```
